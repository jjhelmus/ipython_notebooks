```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pyart
```

Read in the SIGMET file and print out some information on what was read in.

```python
filename = 'JEP100802152831.RAWJY9F'
radar = pyart.io.read(filename)
radar.info('c')
```

Plot the first sweep. Notice that when we set vmax large we pick up the high dBZ artifact near the radar.  This is likely due to ground clutter.

```python
display = pyart.graph.RadarDisplay(radar)
display.plot_ppi('reflectivity', 0, mask_tuple=['reflectivity', 14], vmin=0, vmax=140.)
```

Make a Cartesian grid from the radar data.  The parameter have changed slightly in the new release of Py-ART so that the dimension are specified as z, y, x.   

```python
grid = pyart.map.grid_from_radars(
    (radar,),
    grid_shape=(1, 600, 600),
    grid_limits=((2000, 5000), (-100000.0, 100000.0), (-100000.0, 100000.0)),
    weighting_function='Cressman',
    fields=['reflectivity'])
```

Plot the grid using imshow.  Not the ideal method for plottig but is servers out purpose here.  Notice that the artifact near the radar are also present in the grid.

```python
plt.imshow(grid.fields['reflectivity']['data'][0], vmin=0, vmax=140)
```

Create a new grid that filters out high dBZ reflectivity by setting the `refl_filter_flag` and `max_refl` parameters.  Hopefully this will remove the artifacts near the radar. 

```python
grid_filtered = pyart.map.grid_from_radars(
    (radar,),
    grid_shape=(1, 600, 600),
    grid_limits=((2000, 5000), (-100000.0, 100000.0), (-100000.0, 100000.0)),
    weighting_function='Cressman',
    fields=['reflectivity'],
    refl_filter_flag=True, max_refl=70)
```

Plot the new grid.

```python
plt.imshow(grid_filtered.fields['reflectivity']['data'][0], vmin=0, vmax=140)
```

Masked grid points below 14 dBZ to clean up the image

```python
grid_filtered.fields['reflectivity_masked'] = {
    'data': np.ma.masked_less(grid_filtered.fields['reflectivity']['data'], 14)}
plt.imshow(grid_filtered.fields['reflectivity_masked']['data'][0], vmin=0, vmax=140)
```
